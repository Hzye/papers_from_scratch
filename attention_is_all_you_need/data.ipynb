```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Load and explore data

```python
with open("input.txt", "r", encoding="utf-8") as file:
    text = file.read()
```

```python
len(text)
```

```python
unique_chars_list = sorted(list(set(text)))
unique_chars_str = "".join(unique_chars_list)
unique_chars_str
```

```python
len(unique_chars_str)
```

```python
char_counts = [text.count(char) for char in unique_chars_list]
```

```python
df_chars = pd.DataFrame({"char": unique_chars_list, "counts": char_counts}).sort_values(by=["counts"], ascending=False)
df_chars.head()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15, 15), gridspec_kw={"height_ratios": [1, 3]})
fig.suptitle("Character Frequencies")

ax[0].bar(x=df_chars["char"], height=df_chars["counts"])
ax[0].set_xlabel("Character")
ax[0].set_ylabel("Frequency")
ax[1].pie(df_chars["counts"], labels=df_chars["char"]);
```
